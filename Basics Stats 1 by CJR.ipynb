## Descriptive Analytics and Data Preprocessing on Sales & Discounts Dataset

#### Descriptive Analytics 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
#Loading the dataset
sale_dis=pd.read_csv('sales_data_with_discounts.csv')
print(sale_dis.head())
#Inspecting data
sale_dis.info()
```

```python
#Changing the coloumn names for easibility
sale_dis.rename(columns={
    'Avg Price':'avg_price',
    'Discount Rate (%)':'discount_rate',
    'Total Sales Value':'total_sales_val',
    'Discount Amount':'discount_amt',
    'Net Sales Value':'net_sales_val'
},inplace=True)
```

```python
# Mean Median calculation 
# for volume
print(f'Median :{sale_dis['Volume'].median()}')
print(f'Mode :{sale_dis['Volume'].mode()[0]}')
print(sale_dis['Volume'].describe())
print('='*96)
# for Average price
print(f'Median:{sale_dis['avg_price'].median()}')
print(f'Mode:{sale_dis['avg_price'].mode()[0]}')
print(sale_dis['avg_price'].describe())
print('='*96)
#for total sales value
print(f'Median:{sale_dis['total_sales_val'].median()}')
print(f'Mode:{sale_dis['total_sales_val'].mode()[0]}')
print(sale_dis['total_sales_val'].describe())
print('='*96)
# for Discount Rate
print(f'Median:{sale_dis['discount_rate'].median()}')
print(f'Mode:{sale_dis['discount_rate'].mode()[0]}')
print(sale_dis['discount_rate'].describe())
print('='*96)
# for discount ammount
print(f'Median:{sale_dis['discount_amt'].median()}')
print(f'Mode:{sale_dis['discount_amt'].mode()[0]}')
print(sale_dis['discount_amt'].describe())
print('='*96)
# for Net_sales_value
print(f'Median:{sale_dis['net_sales_val'].median()}')
print(f'Mode:{sale_dis['net_sales_val'].mode()[0]}')
print(sale_dis['net_sales_val'].describe())
print('='*96)
```

### Data Visualization

#### Histogram

```python
# histogram plot for Average price
sns.histplot(sale_dis['avg_price'],kde=True, bins=20, color='#bb8fce')
plt.title('Histogram of Average_Price')
plt.xlabel(' Avg_price')
plt.ylabel("Frequency")
plt.show()
# histogram plot for Total sales value
sns.histplot(sale_dis['total_sales_val'],kde=True, bins=20, color='#c25aef')
plt.title('Histogram of Total_sales_val')
plt.xlabel('total_sales_val')
plt.ylabel("Frequency")
plt.show()
# histogram plot for Discount rate
sns.histplot(sale_dis['discount_rate'],kde=True, bins=20, color='#1abc9c')
plt.title('Histogram of discount_rate')
plt.xlabel('discount_rate')
plt.ylabel("Frequency")
plt.show()
# histogram plot for discount amount
sns.histplot(sale_dis['discount_amt'],kde=True, bins=20, color='#f4d03f')
plt.title('Histogram of discount_amt')
plt.xlabel('discount_amt')
plt.ylabel("Frequency")
plt.show()
# histogram plot for net sales value
sns.histplot(sale_dis['net_sales_val'],kde=True, bins=20, color='#7dcea0')
plt.title('Histogram of Net_sales_val')
plt.xlabel('net_sales_val')
plt.ylabel("Frequency")
plt.show()
# histogram plot for Volume
sns.histplot(sale_dis['Volume'],kde=True, bins=20, color='blue')
plt.title('Histogram of Volume')
plt.xlabel("Volume")
plt.ylabel("Frequency")
plt.show()
```

#### Box Plot

```python
#Box plot for Average price
sns.catplot(data=sale_dis,x='Day',y='avg_price', kind='box',height=3,aspect=2.5,color='#7dcea0')
plt.show()
#Box plot for Total sales  value
sns.catplot(data=sale_dis,x='Day',y='total_sales_val', kind='box',height=3,aspect=2.5,color='#bb8fce')
plt.show()
#Box plot for discount rate
sns.catplot(data=sale_dis,x='Day',y='discount_rate', kind='box',height=4,aspect=3,hue='BU')
plt.show()
#Box plot for discount amount
sns.catplot(data=sale_dis,x='Day',y='discount_amt',kind='box',height=3,aspect=2.5,color='#f4d03f')
plt.show()
#Box plot for net sales value
sns.catplot(data=sale_dis,x='Day',y='net_sales_val', kind='box',height=3,aspect=2.5,color='#85c1e9')
plt.show()
#Box plot for Volume
sns.catplot(data=sale_dis,x='Day',y='Volume',kind='box',height=5,aspect=3,hue='BU')
plt.show()
```

### Standardization of Numerical Variables

```python
standardized_data = sale_dis.copy()
standardized_data['Volume'] = (sale_dis['Volume'] - sale_dis['Volume'].mean()) / sale_dis['Volume'].std()
standardized_data['avg_price']=(sale_dis['avg_price']-sale_dis['avg_price'].mean())/sale_dis['avg_price'].std()
standardized_data['total_sales_val']=(sale_dis['total_sales_val']-sale_dis['total_sales_val'].mean())/sale_dis['total_sales_val'].std()
standardized_data['discount_rate']=(sale_dis['discount_rate']-sale_dis['discount_rate'].mean())/sale_dis['discount_rate'].std()
standardized_data['discount_amt']=(sale_dis['discount_amt']-sale_dis['discount_amt'].mean())/sale_dis['discount_amt'].std()
standardized_data['net_sales_val']=(sale_dis['net_sales_val']-sale_dis['net_sales_val'].mean())/sale_dis['net_sales_val'].std()
standardized_data.head()
```
